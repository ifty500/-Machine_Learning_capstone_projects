When mapping Strings to numbers like this, it is important to make sure that continuous numbers make sense. For example, it wouldn’t make much sense to map "United-States" to 0, "Germany" to 1, and "Mexico" to 2. If we did this, we’re saying that Mexico is more similar to Germany than it is to the United States.

However, if you had values in a column like "low", "medium", and "high" mapping those values to 0, 1, and 2 would make sense because their representation as Strings is also continuous.

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier


income_data = pd.read_csv("income.csv",header= 0,delimiter=", ")

income_data.head()
#print(income_data.head())
```

```python
income_data["country-int"] = income_data["native-country"].apply(lambda x: 0 if x == 'United-States' else 1)
```

```python
#income_data['sex-int'] =income_data['sex'].map({'Male':0, 'Female':1})
income_data['sex-int'] =income_data['sex'].apply(lambda x: 0 if x=="Male"  else 1)
```

```python
income_data['workclass-int'] =income_data['workclass'].apply(lambda x: 0 if x=="Private"  else 1)
```

```python
labels = income_data[['income']]
data = income_data[["age", "capital-gain", "capital-loss", "hours-per-week", "sex-int","country-int","workclass-int","education-num"]]
data
```

```python
train_data, test_data, train_label,test_label=train_test_split(data,labels, random_state=1)
```

```python
scores =[]
for i in range(1,20):
  tree = DecisionTreeClassifier(random_state=1, max_depth=i)

  tree.fit(train_data,train_label)

  score =tree.score(test_data,test_label)

  scores.append(score)
plt.plot(range(1,20), scores)
plt.show()
```

```python
print(max(scores))
```

```python
forest = RandomForestClassifier(random_state=1)

forest.fit(train_data,train_label)
names =data.columns

#print(forest.feature_importances_, name)
print(sorted(zip(map(lambda x: round(x, 4), forest.feature_importances_), names), 
             reverse=True))

```

```python
result=forest.score(test_data, test_label)

print(result)
```
